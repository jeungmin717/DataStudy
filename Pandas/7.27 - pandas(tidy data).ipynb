Wide - > long 포맷으로 변환하는게 분석하기 좋다 .  멜트라는 함수를 쓴다

```python
melt_data = data.melt('religion') # 들어가는 인자가 살릴껏.  melt 를 지원하여 실수를 방지한다. religion 을 기준으로 , 헤드가 -> 변수의 값으로 들어간다.
```

```python
melt_data[melt_data.religion=="Hindu"]
```

불교보다 힌두가 더 분석하기 좋은데이터가 된다 tidy  데이터 .

pandas 분석, aggregation  : groupby , visualization.

```python
melt_data.groupby('religion')
```

groupby는 결과를 중간객체로 바꾼다. 
중간객체뒤에는 대게 통계적 함수가 온다

```python
melt_data.groupby('religion').mean() #이렇게 되면 종교 별 평균 . 데이타가 이런형식이면 다른 함수를 적용하기가 쉬워 좋다 .(tidy data)
```

```python
melt_data.groupby('religion').max()
```

```python
melt_data.groupby('religion').min()
```

```python
data2 = pd.read_csv('tb.csv')
```

```python
data2
```

```python
melt_data2 =data2.melt(['iso2','year']) # 두개를 기준으로 잡을거고 , 한개 이상일때는 리스트!!
```

```python
melt_data2
```

```python
#여기서NaN 은 안중요해보인다. 따라서 drop NaN
melt_data2 =data2.melt(['iso2','year'],value_name="정",var_name='윰') # 두개를 기준으로 잡을거고 , 한개 이상일때는 리스트!!
```

```python
melt_data2
```

```python
melt_data2.dropna(inplace=True)
```

```python
melt_data2
```

```python
melt_data2.groupby(['iso2','year']).sum()  #두개이상 묶으면 리스트로 쓴다. 멀티 인덱스 결과를 총합해서 보여준디 : aggregation  그룹을 지어서 통계값을 내준다 . 
```
