```python
from sklearn.model_selection import train_test_split
```

training set 을 네가지로 나눈다. training set 에 target data 와 학습시킬 data ,

```python
# 학습시킬 데이터는 
# 메트리스 : 관례상 대문자. 
# 타겟은 관례상 소문자.
```

```python
# 이렇게 나눠사하는 방식을 hold_up 방식이라한다 . 
```

```python
#우리는 실제값을 알고 있다 y_test 다 . 
knn.predict(X_test) # predict 값.
```
